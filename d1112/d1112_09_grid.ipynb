```python
import pandas as pd
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
# matplotlib.rcParams['font.family'] = 'AppleGothic'    # Mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # Win용
matplotlib.rcParams['font.size'] = 15  # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx')
```

```python
df.shape
df.shape[0]
# df.count()
df['이름'][0]
```

그래프 grid : 안내선 출력

```python
plt.plot(df['이름'],df['키'])
plt.xticks(rotation=45)

## grid : 그래프 안내선 / axis : x,y축으로 선이 그려짐 
# linestyle : 선 모양 / linewidth : 선 두께 / color : 선 색상
plt.grid(axis='y',color='pink',linestyle='--',linewidth=2,alpha=0.5)
plt.grid(axis='x',alpha=0.5)

# 그래프에 값 추가
for i in range(8):
  plt.text(df['이름'][i],df['키'][i]+0.8,df['키'][i],ha='center') 

plt.show()
```
