```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
# matplotlib.rcParams['font.family'] = 'AppleGothic'    # Mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # Win용
matplotlib.rcParams['font.size'] = 15  # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

그래프에 값 표시 : plt.text

```python
## x,y축의 데이터
x = [1,2,3]  # 리스트, 튜플, DataFrame데이터 다 가능 (리스트형태,Series)
y = [2,4,8]

plt.figure(figsize=(7,5))

plt.plot(x,y,label="범레데이터",marker="o")

## text(x좌표위치값, y좌표위치값, 데이터값) / ha = 글자를 가운데정렬
# plt.text(x[0],y[0]+0.2,y[0],ha='center')
# plt.text(x[1],y[1]+0.2,y[1],ha='center')

## for문 그래프에 값 표시
for idx,txt in enumerate(y):
  plt.text(x[idx]+0.05,y[idx]-0.4,y[idx],ha='center')

plt.title('라인그래프')
plt.legend()
plt.show()
```
