```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
# matplotlib.rcParams['font.family'] = 'AppleGothic'    # Mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # Win용
matplotlib.rcParams['font.size'] = 15  # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
## x,y축의 데이터
x = [1,2,3]  # 리스트, 튜플, DataFrame데이터 다 가능 (리스트 형태, Series)
y = [2,4,8]

## 선그래프 생성
plt.plot(x,y)

## 상단타이틀
plt.title("라인그래프 Line",size=15)  # size : 글자크기 변경

## 그래프 축 설명
# 그래프 x축 설명 / color : 글자색 변경(색상값,16진수코드) / loc : 글자위치(left,center,right)
plt.xlabel('X축 날짜',color='#fa6e66',loc='center')  
# 그래프 y축 설명 / loc : 글자위치(top,center,bottom)
plt.ylabel('Y축 인원',color="green",loc='center')  

## 그래프 눈금단위 표시
plt.xticks([1,2,3])
plt.yticks([2,3,4,8,9,10])

plt.show()
```
