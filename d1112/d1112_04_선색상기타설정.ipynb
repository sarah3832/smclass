```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
# matplotlib.rcParams['font.family'] = 'AppleGothic'    # Mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # Win용
matplotlib.rcParams['font.size'] = 15  # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
## x,y축의 데이터
x = [1,2,3]  # 리스트, 튜플, DataFrame데이터 다 가능 (리스트 형태, Series)
y = [2,4,8]
yy = [1,8,3]

## 선그래프 생성
# linewidth : 선 두께 / color : 선 색상
# marker : 지점표시(o,x,v) / markersize : 마커 크기 / markeredgecolor : 마커 테두리선 색상 / markerfacecolor : 마커 색상
# alpha : 투명도 설정 / linestyle : (:점선, --dsch선, -.선)
plt.plot(x,y,label='인원변동',color='pink',linewidth=5,marker='o',markersize=10,markeredgecolor='black')
# plt.plot(x,yy,label='비용변동',marker='v',markersize=10,markerfacecolor='yellow',linestyle='-.',alpha=0.4)

## 줄이기
# markersize = ms / markerfacecolor = mfc / markeredgecolor = mec / linestyle = ls
# plt.plot(x,yy,label='비용변동',marker='v',ms=10,mfc='green',mec='blue',ls=':',alpha=0.2) 

## 설정별 단축표시 : 'bv:'
# b : 선 색상 / v : 마커표시 / :선 모양
plt.plot(x,yy,'bv:',label='비용변동')

## 상단타이틀
plt.title("라인그래프 Line",size=15)  # size : 글자크기 변경

## 범례 : 그래프 선 설명문구 / loc : 범례 위치 (upper,lower - left,center,right)
# plt.legend(loc='upper center')
plt.legend(loc=(0.3,0.1))  # 좌표값으로 위치설정 가능
```
