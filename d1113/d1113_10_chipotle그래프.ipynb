```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
# matplotlib.rcParams['font.family'] = 'AppleGothic'    # Mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # Win용
matplotlib.rcParams['font.size'] = 15  # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False

df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
df = pd.read_csv('chipotle.tsv',sep='\t')
df.columns
```

```python
df.groupby('item_name')['quantity'].count().mean()
```

```python
## 메뉴별 주문개수
## 순서 상관없이, head(5)로 10개 막대그래프 출력

plt.figure(figsize=(8,3))

menu = df.groupby('item_name')['quantity'].sum().head(10)
# 평균포함
menu['mean'] = df.groupby('item_name')['quantity'].count().mean()

# for i in range(11):
#   plt.text()

plt.grid(axis='y',alpha=0.5)
plt.plot(menu.index,[menu['mean']]*11,ls='--',color='black')
plt.xticks(rotation=90)
bar = plt.bar(menu.index,menu,width=0.6)
bar[len(menu)-1].set_color('red')  # 평균 빨간표시
```
