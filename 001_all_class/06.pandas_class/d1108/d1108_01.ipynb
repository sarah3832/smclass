문자열 함수 : startwith, contains, replace, lower, upper

```python
# startwith : 해당값으로 시작되는 단어 검색
filt = df['이름'].str.startswith('김')
df[filt]

# contains : 해당값이 포함 되어있는 단어 검색
filt = df['이름'].str.contains('수')
df.loc[filt,['이름','학교','키']]

# isin() : 포함이 되어있는지 확인
# isin() : 괄호 안에는 리스트 타입만 가능
langs = ['Python','Java']
filt = df['SW특기'].isin(langs)
df[filt]

# lower() : 해당값을 모두 소문자로 변환
langs = ['python','java']
filt = df['SW특기'].str.lower

# 포함되어 있으면 모두 검색 
# nan데이터 처리를 하지않으면, 에러
filter = df['SW특기'].str.contains('Java',na=False)
df[filt]

# 소문자 j가 포함되어있는 정보검색
filt = df['SW특기'].str.lower().str.contains('j',na=False)
df[filt]
```

```python
## dropna
# axis : 축 방향으로 삭제 / index : row방향 / columns : columns방향
# how : any(1개라도 있으면 모두 row나 columns 삭제) / all : 모두 nan데이터일때 삭제
# axis = 'index', 'columns', how='any', 'all'
df.dropna()  # index, any 기본 default - 해당되는 index(row)를 삭제

df.dropna(axis='index',how='all')  # 모두 nan인 row가 없어서 삭제가 안됨
df.dropna(axis='columns',how='all')  # 학교컬럼이 모두 nan이기에 삭제됨
df
df.dropna(axis='index',how='any')  # 학교컬럼이 모두 nan 이기에 모든 row 삭제
df.dropna(axis='columns',how='any')  # 학교,sw특기 컬럼에 nan있기에 2개컬럼 삭제
```
