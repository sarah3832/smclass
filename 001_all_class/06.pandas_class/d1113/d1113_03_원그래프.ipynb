```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
# matplotlib.rcParams['font.family'] = 'AppleGothic'    # Mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # Win용
matplotlib.rcParams['font.size'] = 15  # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
## 원 그래프 
# values = [30,25,20,13,10,2]
values = [1,1,1,1,2,1]  # 100퍼센트 알아서 맞춰줌
explode = [0.2,0.1,0,0,0,0]
labels = ['python','java','javascript','oracle','html','etc']
colors = ['coral','beige','c','bisque','salmon','tan']
plt.title("프로그램 사용률")

## pie함수(데이터값) / labels : 글자 출력
## explode : 원그래프 분리 / counterclock : 시계 반대방향 (false:시계방향)
## startangle = 90 : 12시부터 시작
## autopct : 데이터 값 출력
## colors : 색상 변경
plt.pie(values,labels=labels,explode=explode,autopct='%.1f%%',counterclock=False,startangle=90,colors=colors)

plt.legend(loc=(1.2,0.3),title='언어별 선호도')
# plt.legend(loc='upper center')
plt.show()
```
