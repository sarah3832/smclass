```python
## 원 그래프 
values = [30,25,20,13,10,2]
# values = [1,1,1,1,2,1]  # 100퍼센트 알아서 맞춰줌
explode = [0.1,0,0,0,0,0]
## edgecolor : 테두리 색상 / linewidth : 테두리 두께
labels = ['python','java','javascript','oracle','html','etc']
colors = ['coral','beige','c','bisque','salmon','tan']
## wedgeprops : 중간공백 (도넛형태)
wedgeprops = {'width':0.6,'edgecolor':'white','linewidth':2}

### 데이터 값이 2% 이하의 값은 출력되지 않도록 함수적용
### pct : values의 퍼센트 값 생성
# def custom_autopect(pct):
#   if pct >= 10:
#     # return ('%.1f%%' % pct)
#     # return ('{:.1f}%'.format(pct))  # format 함수
#     return (f'{pct:.1f}%')  # f 함수
#   else:
#     return ''

## 위와 동일
def custom_autopect(pct):
  return f'{pct:.1f}%' if pct >= 10 else f'{pct:.0f}%'


## wedgeprops : 중간공백 / pctdistance : 글자위치
plt.pie(values,labels=labels,colors=colors,startangle=90,counterclock=False,\
        autopct=custom_autopect,wedgeprops=wedgeprops,pctdistance=0.7)
plt.show()
```
