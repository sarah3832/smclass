```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
# 한글설정
# matplotlib.rcParams['font.family'] = 'AppleGothic'    # Mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # Win용
matplotlib.rcParams['font.size'] = 15  # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False

df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
import pandas as pd
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
df = pd.DataFrame(data)
```

```python
### 평균 막대그래프를 추가하고, 평균 선을(-----) 그어서 표시하시오.
# 이름 : 평균 / 년도 : 2024 / 관객수 : 평균 관객수 / 평점 : 평균 평점
# row 1개를 추가하시오.
df.loc[8] = ['평균',2024,df['관객 수'].mean(),df['평점'].mean()]
```

```python
### 영화 관객수를 막대그래프로 출력하시오.
plt.figure(figsize=(10,4))
data = df.groupby('영화')['관객 수'].sum()
mean = df.groupby('영화')['관객 수'].mean()

con_bar = plt.bar(df['영화'],df['관객 수'],width=0.6)
con_bar[len(data)-1].set_color('red')  # 빨간색 표시

plt.plot(df['영화'],[df['관객 수'][8]]*9,ls='--',color='black')  # 선긋기
plt.text(df['영화'][0],df['관객 수'][0],int(df['관객 수'][0]),ha='center')
plt.xticks(rotation=45)
plt.grid(axis='y',ls='--',alpha=0.5)
plt.ylim(1200,1800)
plt.show()
```
