* fillna : NaN 데이터 해당값으로 채워줌
* drop : NaN 데이터를 삭제
* axis : 축 index-row방향으로 삭제 / colums-culums 방향으로 삭제
* how : 어떤경우에 삭제할지 경우 any-1개라 NaN값이 있으면 삭제, all-모두 NaN일때 삭제

* np.nan : numpy라이브러리 NuN입력

```python
# dropna : NaN 데이터 삭제 
# axis : 축 / index : row삭제 / columns : column 삭제
# how default : any - NaN값이 1개라도 잇으면 삭제 / all - 전부 NaN일때 삭제됨
df.dropna(axis='columns',how=all)
```
