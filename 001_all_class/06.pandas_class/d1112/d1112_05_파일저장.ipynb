```python
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
# matplotlib.rcParams['font.family'] = 'AppleGothic'    # Mac용
matplotlib.rcParams['font.family'] = 'Malgun Gothic'    # Win용
matplotlib.rcParams['font.size'] = 15  # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
import random
df = pd.read_excel('score.xlsx',index_col='지원번호')
```

```python
# x축 : 지원번호 / y축 : 국어점수
# 선 색상:red / marker색상:yellow / 테두리색상:blue / 선타입 : -- 

x = df.index
k = df['국어']
e = df['영어']

# figsize : 그래프 크기조정 / dpi : 그래프 해상도 조정 (높을수록 이미지 크기가 커짐)
# facecolor : 배경색 설정
plt.figure(figsize=(20,5),dpi=200,facecolor='pink')  # x축으로 5만큼, y축으로 20만큼 크기조절

plt.plot(x,k,label='국어',marker='o',ms=15,color='red',mfc='yellow',mec='blue',ls='--')
plt.plot(x,e,label='영어',marker='v',ms=15,color='black',mfc='pink',mec='black',ls=':')


plt.title("성적 그래프",size=15)
plt.xlabel('지원번호')
plt.ylabel('성적')

plt.legend()

# 그래프 파일로 저장
# plt.savefig('graph2.png', dpi=300)
```
