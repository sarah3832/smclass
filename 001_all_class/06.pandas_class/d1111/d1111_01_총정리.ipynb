```python
### 
# 1. 요일 컬럼 추가, 0추가
df['요일'] = 0

# 2. 0-6 까지의 숫자를 랜덤으로 수정
import random

for i in range(8):
  df.iloc[i,9] = random.randint(0,6)

# 3. 함수를 사용해서 0-일요일, 1-월요일, 2-화, .... 6-토요일 변경
def func(value):
  if value == 0:
    return "일요일"
  elif value == 1:
    return "월요일"
  elif value == 2:
    return "화요일"
  elif value == 3:
    return "수요일"
  elif value == 4:
    return "목요일"
  elif value == 5:
    return "금요일"
  elif value == 6:
    return "토요일"

df['요일'] = df['요일'].apply(func)
df
```
